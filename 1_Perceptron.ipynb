```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic data for "Not Placed" (left side)
CGPA_not_placed = np.random.uniform(3.0, 4.0, 50)
IQ_not_placed = np.random.uniform(90, 110, 50)
not_placed = np.column_stack((CGPA_not_placed, IQ_not_placed))

# Generate synthetic data for "Placed" (right side)
CGPA_placed = np.random.uniform(6.0, 9.0, 50)
IQ_placed = np.random.uniform(110, 130, 50)
placed = np.column_stack((CGPA_placed, IQ_placed))
```

```python
# Combine the data
X = np.vstack((not_placed, placed))
y = np.array([0]*50 + [1]*50)

# Create a DataFrame
df = pd.DataFrame(X, columns=['CGPA', 'IQ'])
df['Placement'] = y
```

```python
df.head()
```

```python
df.shape
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='CGPA', y='IQ', hue='Placement', palette=['blue', 'red'])
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Scatter Plot of CGPA vs IQ with Placement as Hue')
plt.show()
```

```python
X1 = df.iloc[:,0:2]
y1 = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron()
p.fit(X1, y1)
```

```python
print(p.coef_)
print(p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X1.values, y1.values, clf=p, legend=2)
```
